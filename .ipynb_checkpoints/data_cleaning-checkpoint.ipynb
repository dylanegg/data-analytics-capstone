```python
import pandas as pd

data = pd.read_csv('infant_mortality_combined_dataset.csv')

data.columns
data.describe()      
```

```python
#check for null values
data.isnull().any()  
```

```python
#check for % of null values
data.isnull().sum()/data.shape[0]
```

```python
#drop all rows with nulls
new_data = data.dropna()
new_data.describe()
```

```python
#ensure all columns are correct data type
new_data.dtypes
```

```python
#drop and non-numeric data types in INFANT_MORTALITY_RATE

new_data['INFANT_MORTALITY_RATE'] = pd.to_numeric(new_data['INFANT_MORTALITY_RATE'], errors = 'coerce')
final_data = new_data.dropna()
final_data.dtypes
```

```python
#check for outliers
import matplotlib.pyplot as plt
final_data.hist()
plt.show()
```

```python
#describe the final data set
final_data.describe()
```

```python
#export the new data to a new csv
final_data.to_csv('final_dataset.csv')
```
